# Data Preprocessing Template

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

pd.read_csv is used to read the csv file.

X is the independent variables in the form of a Numpy matrix. Contains the input features.

y is the dependent variable in the form of a numpy vector. Contains the variable we want to predict

```python
df = pd.read_csv('Data.csv')
X = df.iloc[:, :-1].values          # : means within the range, so takes all the rows, then using -1 means we take all the rows except the last one
y = df.iloc[:, -1].values              

#iloc is used to index the columns
#values used to return the values of the columns inside the numpy arrays. How X and y become numpy arrays
```

## Splitting the dataset into the Training set and Test set

Training set is used to actually train the model, and testing set is used to evaluate the model on new data not trained with the model

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```
